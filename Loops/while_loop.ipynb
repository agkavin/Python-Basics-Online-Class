# What is While Loop

With the while loop we can execute a set of statements as long as a condition is true.

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

### break statement with while loop

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

### continue statement with while loop

```python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```

### When to use which Loop ?

For loop is used when the number of iterations is determined.
and while loop used when the range is not known.

Example:

```python
for i in range(5):
    print(i)
```

```python
scores = [85, 92, 76, 58, 99, 67]
grades = []

for score in scores:
    if score >= 90:
        grades.append('A')
    elif score >= 80:
        grades.append('B')
    elif score >= 70:
        grades.append('C')
    elif score >= 60:
        grades.append('D')
    else:
        grades.append('F')

print(grades)
```

```python
password = ''
while password != 'secret':
    password = input("Enter password: ")

print("Access granted")
```

```python
import random

readings = []
reading = 0
count = 0


while reading < 90:
    reading = random.randint(0, 100)
    readings.append(reading)
    print(f"Sensor reading:{reading}")
    count += 1

print(f"\nThreshold reached after {count} readings.")
print(f"All readings: {readings}")
```

### while else

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
else:
  print("i is no longer less than 6")
```

### Infinite Loop

```python
i = 1
while i < 5:
    print(i)
    i += 1
```

### Fibonacci sequence

```python
n = int(input("Enter a number: "))
sequence = []
a, b = 0, 1
while a <= n:
    sequence.append(a)
    a, b = b, a + b
print(sequence)
```

### Russian Roulette

```python
import random

print("Welcome to Russian Roulette!")
print("There is 1 bullet in a 6-chamber revolver.")

chambers = [0, 0, 0, 0, 0, 1]
random.shuffle(chambers)
position = 0
dead = False

while not dead:
    input("Press Enter to pull the trigger...")
    print("You pull the trigger...")

    if chambers[position] == 1:
        print(" Bang! You're dead.")
        dead = True
    else:
        print(" Click. You're still alive.")
        position += 1
        if position > 5:
            print("The cylinder is empty. You survived all 6 shots!")
            break
```

### Binary Search

```python
arr = [1, 3, 5, 7, 9, 11]
target = 11
left = 0
right = len(arr) - 1
cnt = 0

while left <= right:
    cnt += 1
    mid = (left + right) // 2
    
    if arr[mid] == target:
        print(mid)
        break

    elif arr[mid] < target:
        left = mid + 1
    else:
        right = mid - 1

else:
    print(-1)

print("count:", cnt)
```

### Linear Search

```python
arr = [1, 3, 5, 7, 9, 11]
t = 11
cnt = 0


for i in range(len(arr)):
    if arr[i] == t:
        print(i)
        break
    else:
        cnt += 1

else:
    print(-1)

print("count:", cnt)
```
